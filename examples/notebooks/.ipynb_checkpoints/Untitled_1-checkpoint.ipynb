```python
from polypy import read_dl_poly as dlpy
from polypy import read as rd
import numpy as np
from polypy import trajectory
```

```python
traj = dlpy.read_trajectory("../example_data/HISTORY")

traj_obj = trajectory.Trajectory(traj)

lvs = traj_obj.lattice_vectors()
```

```python
print(lvs)
```

```python
natoms = traj_obj.get_natoms()
```

```python
print(traj[0]['timestep'])
print(traj[0]['lvs'])
print(traj[0]['atoms']['numatoms'])
for i in range(traj[0]['atoms']['numatoms']):
    print(traj[0]['atoms'][i])


    
coords = []
for iconfig in range( traj['numconfigs']):
    this_config = traj[iconfig]
    for iatom in range(this_config['atoms']['numatoms']):
        coords.append(this_config['atoms'][iatom]['coor'])

coords = np.asarray(coords)
print(coords.size)
print(coords.shape)
```
